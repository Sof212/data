```python
# -----------------------------
# 3ï¸âƒ£ CrÃ©er un BinaryLabelDataset
# -----------------------------
dataset = BinaryLabelDataset(
    df=data,
    label_names=["target"],
    protected_attribute_names=["sex", "race"]
)

# -----------------------------
# 4ï¸âƒ£ Calculer les metrics
# -----------------------------
def compute_metrics(dataset, protected_attr, privileged_val, unprivileged_val):
    metric = BinaryLabelDatasetMetric(
        dataset,
        privileged_groups=[{protected_attr: privileged_val}],
        unprivileged_groups=[{protected_attr: unprivileged_val}]
    )
    return metric.disparate_impact(), metric.mean_difference()

# Sexe : Male(1) vs Female(0)
sex_DI, sex_SPD = compute_metrics(dataset, "sex", 1, 0)

# Race : White(4) vs Black(1)  (selon LabelEncoder)
race_black_DI, race_black_SPD = compute_metrics(dataset, "race", 4, 1)

# Race : White(4) vs Other (tout sauf White=4) => crÃ©er un masque
other_mask = ~data["race"].isin([4])
data_other = data.copy()
data_other.loc[other_mask, "race"] = 0  # 0 = unprivileged group
dataset_other = BinaryLabelDataset(df=data_other, label_names=["target"], protected_attribute_names=["race"])
race_other_DI, race_other_SPD = compute_metrics(dataset_other, "race", 4, 0)

# -----------------------------
# 5ï¸âƒ£ CrÃ©er un DataFrame pour le barplot
# -----------------------------
metrics_df = pd.DataFrame({
    "Biais": ["Sex Male vs Female", "Race White vs Black", "Race White vs Other"],
    "Disparate Impact": [sex_DI, race_black_DI, race_other_DI],
    "Statistical Parity Difference": [sex_SPD, race_black_SPD, race_other_SPD]
})

# -----------------------------
# 6ï¸âƒ£ Barplot
# -----------------------------
plt.figure(figsize=(10,6))
metrics_df_melted = metrics_df.melt(id_vars="Biais", var_name="Metric", value_name="Value")
sns.barplot(x="Biais", y="Value", hue="Metric", data=metrics_df_melted)
plt.title("Comparaison des biais dans le dataset Adult")
plt.ylabel("Valeur de la metric")
plt.xticks(rotation=15)
plt.show()
```

Bien sÃ»r ðŸ˜„ voici deux paragraphesâ€¯:

---

### Paragraphe simple (explication intuitive)

Le **Disparate Impact Remover (DIR)** est un outil pour rÃ©duire le biais dans les donnÃ©es avant dâ€™entraÃ®ner un modÃ¨le. Il regarde les colonnes continues et ajuste leurs valeurs afin que les groupes privilÃ©giÃ©s et non-privilÃ©giÃ©s aient des distributions plus similaires, tout en gardant autant dâ€™information que possible. Le **reweighing**, quant Ã  lui, ne modifie pas les donnÃ©es mais attribue des **poids diffÃ©rents aux exemples** selon leur groupe et leur label. Lâ€™objectif est que, pendant lâ€™entraÃ®nement, le modÃ¨le ne favorise pas le groupe privilÃ©giÃ© simplement parce quâ€™il est surreprÃ©sentÃ© ou a des labels plus frÃ©quents.

---

### Paragraphe mathÃ©matique

Pour le **Disparate Impact Remover**, on considÃ¨re une feature continue (X) et un attribut sensible (S \in {0,1}) (0 = non-privilÃ©giÃ©, 1 = privilÃ©giÃ©). On cherche une transformation (f(X,S)) telle que la **distribution conditionnelle** de la feature soit proche pour les deux groupes :

[
f(X \mid S=0) \approx f(X \mid S=1)
]

Typiquement, DIR utilise une **transformation affine ou monotone** basÃ©e sur les **quantiles** pour aligner les distributions tout en gardant la structure relative des donnÃ©es.

Pour le **reweighing**, on calcule un poids (w(x,s,y)) pour chaque exemple ((x,s,y)) selon la formule :

[
w(x,s,y) = \frac{P(S=s) , P(Y=y)}{P(S=s, Y=y)}
]

oÃ¹ (P(S=s)) et (P(Y=y)) sont les probabilitÃ©s marginales, et (P(S=s,Y=y)) la probabilitÃ© conjointe. Ces poids permettent dâ€™ajuster la contribution de chaque exemple lors de lâ€™entraÃ®nement pour que le modÃ¨le traite tous les groupes de maniÃ¨re Ã©quitable.

Bien sÃ»r ðŸ˜„ dÃ©taillons la formule du **reweighing**.

---

### Rappel de la formule

Pour chaque exemple ((x, s, y)) dans le dataset :

[
w(x,s,y) = \frac{P(S=s) \cdot P(Y=y)}{P(S=s, Y=y)}
]

oÃ¹ :

* (S) = attribut sensible (ex. sexe : Male/Female)
* (Y) = label ou classe (ex. revenu >50K ou â‰¤50K)
* (x) = les features du reste des donnÃ©es

---

### InterprÃ©tation Ã©tape par Ã©tape

1. **ProbabilitÃ©s marginales**

   * (P(S=s)) = fraction des exemples du groupe (s) dans tout le dataset
   * (P(Y=y)) = fraction des exemples avec le label (y) dans tout le dataset

2. **ProbabilitÃ© conjointe**

   * (P(S=s, Y=y)) = fraction des exemples qui sont Ã  la fois dans le groupe (s) **et** ont le label (y)

3. **Pourquoi on divise par la probabilitÃ© conjointe ?**

   * Si un certain groupe (s) avec un label (y) est **sous-reprÃ©sentÃ©**, (P(S=s, Y=y)) sera petite
   * En divisant (P(S=s) \cdot P(Y=y)) par (P(S=s, Y=y)), on **augmente le poids** de ces exemples sous-reprÃ©sentÃ©s
   * Inversement, si un groupe est sur-reprÃ©sentÃ©, le poids sera rÃ©duit

4. **Effet global**

   * AprÃ¨s avoir appliquÃ© ces poids pendant lâ€™entraÃ®nement, le modÃ¨le **voit chaque combinaison groupe-label de maniÃ¨re â€œÃ©quilibrÃ©eâ€**
   * Cela rÃ©duit le biais qui pourrait survenir si certaines combinaisons sont surreprÃ©sentÃ©es dans le dataset

---

### Exemple simple

Supposons un dataset de 100 personnes :

| Groupe (S) | Label (Y) | Count |
| ---------- | --------- | ----- |
| Male       | 1         | 40    |
| Male       | 0         | 10    |
| Female     | 1         | 5     |
| Female     | 0         | 45    |

* (P(S=\text{Male}) = 50/100 = 0.5)
* (P(Y=1) = 45/100 = 0.45)
* (P(S=\text{Male}, Y=1) = 40/100 = 0.4)

Le poids pour un homme avec label 1 :

[
w = \frac{0.5 \cdot 0.45}{0.4} = 0.5625
]

* Pour une femme avec label 1 :

[
w = \frac{0.5 \cdot 0.45}{0.05} = 4.5
]

âœ… Comme on voit, les femmes avec label 1 Ã©taient sous-reprÃ©sentÃ©es â†’ leur poids augmente fortement.
