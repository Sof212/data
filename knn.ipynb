```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# Charger les donn√©es
data_train = pd.read_csv("/kaggle/input/projet-data/train.csv")
data_test = pd.read_csv("/kaggle/input/projet-data/test.csv")

# üöÄ Pr√©parer les donn√©es d'entra√Ænement
X_train, y_train = [], []

# Cr√©er les donn√©es pour l'entra√Ænement avec les indices des valeurs manquantes comme features
for index, row in data_train.iterrows():
    known_indices = np.where(~np.isnan(row))[0]  # Indices des valeurs connues
    X_train.extend(known_indices.reshape(-1, 1))  # Ajouter les indices des valeurs connues
    y_train.extend(row[known_indices])  # Ajouter les valeurs connues

# Convertir en array numpy pour entra√Æner le mod√®le
X_train = np.array(X_train)
y_train = np.array(y_train)

# Entra√Æner un Random Forest sur toutes les s√©ries
rf = RandomForestRegressor(n_estimators=1, random_state=42)
rf.fit(X_train, y_train)
i=0
# üöÄ Imputer les donn√©es manquantes dans le dataset test
data_test_imputed = data_test.copy()
for index, row in data_test.iterrows():
    print(i)
    test_series = np.array(row)  # S√©rie temporelle du test
    missing_indices = np.where(np.isnan(test_series))[0]  # Indices des valeurs manquantes

    if len(missing_indices) > 0:
        # Pr√©dire les valeurs manquantes
        X_missing = missing_indices.reshape(-1, 1)  # Les indices des valeurs manquantes
        y_pred = rf.predict(X_missing)  # Pr√©dictions

        # Remplacer les valeurs manquantes dans la s√©rie
        test_series[missing_indices] = y_pred
        data_test_imputed.iloc[index] = test_series  # Mettre √† jour la ligne dans le DataFrame
        i=i+1
# ‚úÖ R√©sultat apr√®s imputation
print(data_test_imputed)
```

```python
idx_with_nan = data_test.isna().values.flatten()
pred =  data_test_imputed.values.flatten()[idx_with_nan]
submission_example = pd.DataFrame(
    data={'ID': [i for i in range(len(pred))],
          'TARGET': pred
          }
    )
print(submission_example.to_csv('rdforest1.csv', index=False))
print(submission_example.isnull().values.any())
pred.shape
```

```python
import numpy as np
import pandas as pd
from catboost import CatBoostRegressor
from sklearn.preprocessing import StandardScaler

# Charger les donn√©es
train = pd.read_csv("/kaggle/input/projet-data/train.csv")
test = pd.read_csv("/kaggle/input/projet-data/test.csv")

# Cr√©er une copie pour l'imputation
data_test_imputed = test.copy()

# Normalisation des donn√©es (optionnel, d√©pend du dataset)
scaler = StandardScaler()
data_combined = pd.concat([train, test], ignore_index=True)
data_combined_scaled = scaler.fit_transform(data_combined)

# Appliquer CatBoost pour chaque ligne avec des valeurs manquantes
for index, row in test.iterrows():
    # Extraire la s√©rie temporelle compl√®te
    train_series = np.array(train.iloc[index])
    test_series = np.array(row)
    
    # Trouver les indices des valeurs connues et manquantes
    known_indices = np.where(~np.isnan(test_series))[0]
    missing_indices = np.where(np.isnan(test_series))[0]
    
    if len(missing_indices) > 0:
        # Pr√©parer les donn√©es d'entra√Ænement avec les indices des valeurs connues
        X_train = known_indices.reshape(-1, 1)  # Indices des valeurs connues
        y_train = train_series[known_indices]   # Valeurs correspondantes dans train

        # Initialiser le mod√®le CatBoostRegressor
        model = CatBoostRegressor(iterations=100, learning_rate=0.05, depth=6, cat_features=[], verbose=0)
        
        # Entra√Æner le mod√®le sur les donn√©es connues
        model.fit(X_train, y_train)

        # Pr√©dire les valeurs manquantes
        X_missing = missing_indices.reshape(-1, 1)
        y_pred = model.predict(X_missing)

        # Remplir les valeurs manquantes
        test_series[missing_indices] = y_pred
        data_test_imputed.iloc[index] = test_series

# R√©sultat apr√®s imputation
print(data_test_imputed)

# Sauvegarder les r√©sultats dans un fichier CSV
data_test_imputed.to_csv('catboost_imputed.csv', index=False)
```

```python
idx_with_nan = data_test.isna().values.flatten()
pred =  data_test_imputed.values.flatten()[idx_with_nan]
submission_example = pd.DataFrame(
    data={'ID': [i for i in range(len(pred))],
          'TARGET': pred
          }
    )
print(submission_example.to_csv('catboost100.csv', index=False))
print(submission_example.isnull().values.any())
pred.shape
```
